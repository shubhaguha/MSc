```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def _read_data_file(filename):
    train_stats = []
    valid_stats = []
    num_updates = 0
    for line in open(filename, 'r'):
        if line.startswith('INFO:'):
            line = line[6:]
        if line.startswith('Epoch'):
            data = line.split()
            num_updates = int(data[3])
            train_cost = float(data[5])
            train_stats += [(num_updates, train_cost)]
        elif line.startswith('Valid'):
            valid_cost = float(line.split()[-1])
            valid_stats += [(num_updates, valid_cost)]
    return train_stats, valid_stats
```

```python
def plot_training_curve(model_name):
    """Plot the change in the validation and training set cost over training."""
    t, v = _read_data_file('{}.txt'.format(model_name))

    fig_1 = plt.figure(figsize=(8, 4))
    ax_1 = fig_1.add_subplot(111)
    ax_1.plot(*zip(*t), label='train')
    ax_1.plot(*zip(*v), label='valid')
    ax_1.legend(loc=0)
    ax_1.set_xlabel('Number of Minibatches')
    ax_1.set_ylabel('Cost')
    ax_1.set_title('Training Curve for {}'.format(model_name))
    fig_1.savefig('{}.pdf'.format(model_name))
```

```python
plot_training_curve('baseline_1')
```
